```python
# Decision Tree using matplotlib (ML Lab)

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder

# --------------------------------------------------
# Step 1: Dataset
# --------------------------------------------------
data = {
    'Age': ['Youth','Youth','Middle','Senior','Senior',
            'Senior','Middle','Youth','Youth','Senior'],
    'Income': ['High','High','High','Medium','Low',
               'Low','Low','Medium','Low','Medium'],
    'Student': ['No','No','No','No','Yes',
                'Yes','Yes','No','Yes','Yes'],
    'Credit_Rating': ['Fair','Excellent','Fair','Fair','Fair',
                      'Excellent','Excellent','Fair','Fair','Fair'],
    'Buys_Computer': ['No','No','Yes','Yes','Yes',
                      'No','Yes','No','Yes','Yes']
}

df = pd.DataFrame(data)

# --------------------------------------------------
# Step 2: Label Encoding
# --------------------------------------------------
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])

# --------------------------------------------------
# Step 3: Feature and Target Split
# IMPORTANT: Keep Age as FIRST column
# --------------------------------------------------
X = df[['Age', 'Student', 'Credit_Rating', 'Income']]
y = df['Buys_Computer']

# --------------------------------------------------
# Step 4: Train Decision Tree (Entropy)
# --------------------------------------------------
model = DecisionTreeClassifier(
    criterion='entropy',
    max_depth=3
)
model.fit(X, y)

print("Decision Tree Model Trained Successfully")

# --------------------------------------------------
# Step 5: Classify New Sample
# Youth, Medium, Yes, Fair
# --------------------------------------------------
new_sample = pd.DataFrame({
    'Age': [2],            # Youth
    'Student': [1],        # Yes
    'Credit_Rating': [1],  # Fair
    'Income': [2]          # Medium
})

prediction = model.predict(new_sample)

print("\nNew Sample Classification:")
print("Buys Computer =", "YES" if prediction[0] == 1 else "NO")

# --------------------------------------------------
# Step 6: Plot Decision Tree using matplotlib
# --------------------------------------------------
plt.figure(figsize=(14, 8))
plot_tree(
    model,
    feature_names=X.columns,
    class_names=['No', 'Yes'],
    filled=True,
    rounded=True
)
plt.title("Decision Tree for Buy Computer Dataset")
plt.show()
```
