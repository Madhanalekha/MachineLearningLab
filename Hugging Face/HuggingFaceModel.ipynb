```python
from transformers import pipeline

# Load pretrained sentiment analysis model
classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

# Test sentences (Spam / Not Spam)
test_sentences = [
    
    "Win money now",
    "Claim your free prize",
    "How are you doing",
    "Lunch at office"
]

# Prediction
for text in test_sentences:
    result = classifier(text)[0]  # returns a dict with label and score
    # Map sentiment label to Spam / Not Spam
    spam_label = "Spam" if result['label'] == "POSITIVE" else "Not Spam"

    print("Text:", text)
    print("Prediction:", spam_label)
    print("Confidence Score:", round(result['score'], 2))
    print("-----------------------------")
```
