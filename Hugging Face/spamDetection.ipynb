```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB

# Training data
train_messages = [
    "win prize now",
    "claim free reward",
    "win money free",
    "prize money win",
    "free prize claim",
    "meeting scheduled today",
    "project meeting tomorrow",
    "submit assignment today",
    "discussion on project",
    "meeting about assignment"
]

train_labels = [
    "Spam", "Spam", "Spam", "Spam", "Spam",
    "Not Spam", "Not Spam", "Not Spam", "Not Spam", "Not Spam"
]

# Convert text to feature vectors
vectorizer = CountVectorizer()
X_train = vectorizer.fit_transform(train_messages)

# Train Naive Bayes model
model = MultinomialNB()
model.fit(X_train, train_labels)

# Test sentences
test_sentences = [
    "Win money now",
    "Claim your free prize",
    "How are you doing",
    "Lunch at office"
]

# Transform test data
X_test = vectorizer.transform(test_sentences)

# Predictions
predictions = model.predict(X_test)
probabilities = model.predict_proba(X_test)

# Output results
for i, text in enumerate(test_sentences):
    print("Text:", text)
    print("Prediction:", predictions[i])
    print("Probability:", round(max(probabilities[i]), 2))
    print("-----------------------------")
```
